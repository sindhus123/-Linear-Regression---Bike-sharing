## Data Cleaning
   1. Removing the date column as we have separate columns for year & month
   2. Removing the instance column as it is the index & of no use in analysis
   3. Removing casual & registered as our target variable is the total count of bike & not based on the user segment

```python
df.columns
```

```python
df = df.drop(['instant','dteday','casual','registered'], axis = 1)
```

```python
df.columns
```

   #### Observation:
   1. Season, mnth, weekday & weathersit are categorical variables.Convering them as category dtype
   2. Replacing them with respective names for easy undersatding 
   3. Yr, Holiday & working day are categorical but has only 2 values 0 or 1 (Yes or No). Its easy to understand.

```python
df.season.replace({1:"spring", 2:"summer", 3:"fall", 4:"winter"},inplace = True)
df.mnth.replace({1:"Jan", 2:"Feb", 3:"Mar", 4:"Apr",5:"May", 6:"Jun", 7:"Jul", 8:"Aug",9:"Sep", 10:"Oct", 11:"Nov", 12:"Dec"},inplace = True)
df.weekday.replace({0:"Sun",1:"Mon", 2:"Tue", 3:"Wed", 4:"Thu",5:"Fri", 6:"Sat"},inplace = True)
df.weathersit.replace({1:"Clear", 2:"Cloudy", 3:"Moderate rain", 4:"Heavy Rain"},inplace = True)
```

```python
df['season'] = df['season'].astype('category')
df['mnth'] = df['mnth'].astype('category')
df['weekday'] = df['weekday'].astype('category')
df['weathersit'] = df['weathersit'].astype('category')
```

## Preparing data for Linear Regression
   #### Creating Dummy variables for categorical data

```python
df_new = pd.get_dummies(df, drop_first = True)
```

```python
df_new.info()
```

```python
df_new.shape
```

```python
plt.figure(figsize = (20,10))
sns.heatmap(df_new.corr(),annot = True, cmap = "YlGnBu")
plt.show()
```

#### Splitting data to train & test data sets

```python
from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split(df_new, train_size = 0.7, test_size = 0.3, random_state = 100)
```

#### Scaling the variables
 Applying scaling to all numeric variables expect dummy variables & yes/no variables

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
df_train.columns
```

```python
num_vars = ['temp','atemp','hum','windspeed','cnt']
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
df_train.describe()
```

#### Dividing into x & y sets

```python
y_train = df_train.pop('cnt')
x_train = df_train
```

## Building the Model

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(x_train,y_train)
```

#### RFE
- We will use RFE to start with 15 variables & have mixed approach in selecting variables

```python
rfe = RFE(lm,15)
rfe = rfe.fit(x_train, y_train)
```

```python
list(zip(x_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = x_train.columns[rfe.support_]
col
```

```python
x_train.columns[~rfe.support_]
```

```python
x_train_lm1 = x_train[col]
```

## Model 1

#### Building model using stats model for detailed satistics 

```python
import statsmodels.api as sm
x_train_lm1 = sm.add_constant(x_train_lm1)
lm1 = sm.OLS(y_train, x_train_lm1).fit()
lm1.summary()
```

#### VIF

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
X = x_train_lm1
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

 #### Observation:
   1. Feature 'mnth_Nov' has high p-value (0.089) so dropping it

## Model 2

```python
x_train_lm2 = x_train_lm1.drop('mnth_Nov',axis = 1)
x_train_lm2 = sm.add_constant(x_train_lm2)
lm2 = sm.OLS(y_train,x_train_lm2).fit()
lm2.summary()
```

#### VIF - Model 2

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
X = x_train_lm2
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

## Residual Analysis

```python
y_train_pred = lm2.predict(x_train_lm2)
```

```python
ax = sns.regplot(x = y_train, y = y_train_pred).set(title = "Y-Train Vs Y-Predicted",xlabel ='y_train',ylabel = 'y_pred')
plt.show()
```

```python
sns.distplot((y_train - y_train_pred), bins = 20).set_title("Error Terms")
plt.show()
```

```python
mean_residual = np.mean(y_train - y_train_pred)
mean_residual
```

Error terms are normally distributed & mean close to zero

```python
ax = sns.regplot(x = y_train_pred, y = (y_train - y_train_pred)).set(title = "Y-Predicted Vs Residual",xlabel ='y_train_pred',ylabel = 'Residual')
plt.show()
```

Error terms have constant variance

```python
sm.graphics.tsa.plot_acf((y_train - y_train_pred), lags =40)
plt.show()
```

Error terms are independant of each other

## Predictions

Applying scaling on test sets

```python
num_vars = ['temp','atemp','hum','windspeed','cnt']
df_test[num_vars] = scaler.transform(df_test[num_vars])
```

Dividing into x & y sets

```python
y_test = df_test.pop('cnt')
x_test = df_test
```

```python
x_train_lm2 = x_train_lm2.drop(['const'], axis =1)
```

Predicting the target

```python
x_test_lm = x_test[x_train_lm2.columns]
x_test_lm = sm.add_constant(x_test_lm)
```

```python
y_test_pred = lm2.predict(x_test_lm)
```

## Evaluation

Y-test vs Y-prediction

```python
ax = sns.regplot(x = y_test, y = y_test_pred).set(title = "Y-Test Vs Y-Predicted",xlabel ='y_test',ylabel = 'y_pred')
plt.show()
```

```python
sns.distplot((y_test - y_test_pred), bins = 20).set_title("Error Terms")
plt.show()
```

```python
mean_residual = np.mean(y_test - y_test_pred)
mean_residual
```

#### Error Terms are normally distributed & mean close to zero

```python
ax = sns.regplot(x = y_test_pred, y = (y_test - y_test_pred)).set(title = "Y-Predicted Vs Residual",xlabel ='y_test_pred',ylabel = 'Residual')
plt.show()
```

Residuals vs fitted value shows error term variance is constant (homoscedasticity)

```python
sm.graphics.tsa.plot_acf((y_test - y_test_pred), lags =40)
plt.show()
```

Error terms are not dependent of each other

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_test_pred)
r2
```

```python
n = x_test.shape[0]
p = x_test.shape[1]
adj_r2 = 1 - (((1-r2)*(n-1))/(n-p-1))
adj_r2
```

### Model Interpretation:
  1. Model 2 - all the p values are <0.05 & hence the coefficients are significant
  2. Temp has the highest co-efficient value of 0.57. This is aligned with the strong correlation between 'cnt' & 'temp'
  3. Next to 'temp', weather(0.25) & Year (0.23) are important co efficients in predicting the values
  4. All the coefficients have value (not equal to zero). Null hypothesis rejected
  5. P(F- static) is very less which explains fit is not by chance
  6. R Square is 84.5% & Adjusted R Square is 84% which explains the variance of the model
  7. Constant (intercept) is 0.284.So without any other predictor value we will have increase of 0.284 units

### Final results:

1. Train Data: R Square = 0.845 ; Adjusted R Square = 0.841
2. Test Data: R Square = 0.819 ; Adjusted R Square = 0.791

The model fits well with the test data and can be used for prediction across various data set.
